```python
import tiktoken

def num_tokens_from_string(string: str, encoding_name: str) -> int:
    """Returns the number of tokens in a text string."""
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens

text = ""

num_tokens = num_tokens_from_string(text, "gpt2")
print(num_tokens)
```

```python
import tkinter as tk
from tkinter import messagebox
import openai
from fpdf import FPDF
import time

def generate_study_notes(input_text):
import os
    openai.api_key = 'os.getenv("OPENAI_API_KEY")'  # Replace with your OpenAI API key

    response = openai.Completion.create(
        engine='text-davinci-003',  # Choose the appropriate language model
        prompt=input_text,
        temperature=0.2,
        max_tokens=1000,
        n=1,  # Number of study note variations to generate
        stop=None,
    )

    study_notes = [note['text'].strip() for note in response['choices']]
    return study_notes

def generate_pdf():
    global start_time
    start_time = time.time()  # Start the timer

    subject = subject_entry.get()
    sub_topic = sub_topic_entry.get()
    topics = topics_entry.get()

    l = topics.split(",")

    # Create PDF object
    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)

    # Set style and size of font
    pdf.add_page()
    pdf.set_font("Arial", size=15)

    # Set the width and height for the MultiCell
    width = 180
    height = 10

    pdf.multi_cell(w=width, h=height, align='C', txt=subject)
    pdf.set_font("Arial", size=12)
    pdf.multi_cell(w=width, h=height, align='L', txt=sub_topic + " :")
    pdf.set_font("Arial", size=10)

    for i in range(len(l)):
        pdf.multi_cell(w=width, h=height, align='L', txt=l[i] + " :")
        input_text = f"in detail explain what is {l[i]} in {sub_topic} in {subject} in detail.give output only in codec which is pdf compatible"
        study_notes = generate_study_notes(input_text)
        # Print the generated study notes
        for note in study_notes:
            encoded_note = note.encode('latin-1', 'ignore').decode('latin-1')
            pdf.multi_cell(w=width, h=height, align='L', txt=encoded_note)
            pdf.ln(1)

    pdf.output("Hackathon.pdf")
    elapsed_time = int(time.time() - start_time)
    messagebox.showinfo("PDF Generation", f"PDF generated successfully. Elapsed Time: {elapsed_time} seconds.")

def update_time():
    if start_time is not None:
        elapsed_time = int(time.time() - start_time)
        timer_label.config(text=f"Elapsed Time: {elapsed_time} seconds")
    window.after(1000, update_time)

# Create the GUI window
window = tk.Tk()
window.title("Study Notes Generator")
window.geometry("400x300")

# Create input fields and labels
subject_label = tk.Label(window, text="Subject:")
subject_label.pack()
subject_entry = tk.Entry(window)
subject_entry.pack()

sub_topic_label = tk.Label(window, text="Subtopic:")
sub_topic_label.pack()
sub_topic_entry = tk.Entry(window)
sub_topic_entry.pack()

topics_label = tk.Label(window, text="Topics (separated by comma):")
topics_label.pack()
topics_entry = tk.Entry(window)
topics_entry.pack()

generate_button = tk.Button(window, text="Generate PDF", command=generate_pdf)
generate_button.pack()

timer_label = tk.Label(window, text="Elapsed Time: 0 seconds")
timer_label.pack()

start_time = None

window.after(1000, update_time)
window.mainloop()
```
